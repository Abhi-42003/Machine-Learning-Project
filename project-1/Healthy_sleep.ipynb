```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\abhij\OneDrive\Desktop\Data science\Notes\health_data\Health_Sleep_Statistics.csv")
print(df.to_string())
```

```python
df=df.rename(columns={'Wake-up Time':'Wake up time'})
df
```

```python
df.columns=[i.replace(' ','_')for i in df.columns]
df
```

```python
#1. How does age affect sleep quality?
df.plot.scatter(x='Age',y='Sleep_Quality')
```

```python
#2. Is there a difference in sleep quality between males and females?
sns.barplot(data = df , x= 'Gender' ,y = 'Sleep_Quality',palette='Set2')
plt.title('Mean Sleep Quality')
plt.show()
```

```python
#3. What is the distribution of daily steps among individuals with different physical activity levels?
sns.boxplot(data=df, x='Physical_Activity_Level', y='Daily_Steps')
plt.title('Distribution of Daily Steps by Physical Activity Level')
plt.show()
```

```python
#5. How does bedtime relate to wake-up time?
df.plot.scatter(x='Bedtime',y='Wake_up_time')
plt.xticks(rotation=45)
plt.show()


```

```python
#6. What is the relationship between physical activity level and sleep quality?
sns.barplot(data=df,x='Physical_Activity_Level',y='Sleep_Quality',palette='hls')
plt.show()
```

```python
df
```

```python
#7. Do individuals with sleep disorders have lower sleep quality?
df.boxplot(column='Sleep_Quality',by='Sleep_Disorders')

```

```python

#8. How do dietary habits affect physical activity levels?
cross_tab=pd.crosstab(df['Dietary_Habits'], df['Physical_Activity_Level']).plot(kind='bar',stacked=False)
```

```python
#9. How does medication usage for sleep disorders vary by gender?
crosstab_med_use = pd.crosstab(df['Gender'], df['Medication_Usage'])
crosstab_med_use.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])
plt.title('Medication Usage for Sleep Disorders by Gender')
plt.ylabel('Count of Individuals')
plt.xlabel('Gender')
plt.show()
```

```python
#10. What is the distribution of sleep quality across individuals with different dietary habits?
sns.barplot(data=df, x='Dietary_Habits', y='Sleep_Quality')
plt.title('Distribution of Sleep Quality by Dietary Habits')
plt.xlabel('Dietary Habits')
plt.ylabel('Sleep Quality')
plt.show()
```
